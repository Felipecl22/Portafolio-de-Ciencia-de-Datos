Asignación de puntos a centroides:

Para cada punto en el conjunto de datos, calculamos la distancia entre ese punto y todos los centroides. Luego, asignamos ese punto al centroide más cercano. Este proceso se repite para todos los puntos en tu conjunto de datos.

Recálculo de los centroides:

Después de asignar todos los puntos a los centroides, recalculamos los centroides. Para cada centroide, calculamos el promedio de las coordenadas de todos los puntos asignados a ese centroide. Esto nos da los nuevos valores de los centroides.

Iteración 2:

Después de la asignación de puntos a centroides en la primera iteración, recalculamos los centroides como el promedio de los puntos asignados a cada uno.

Promedio de las coordenadas z: (31 + 86 + 75 + 14 + 40) / 5 = 49.2
Entonces, el nuevo centroide C2 es (47.2, 55, 49.2).


Repetición del proceso:

Después de recalcular los centroides, volvemos al paso 2 y repetimos el proceso de asignación de puntos a centroides utilizando los nuevos centroides. Este proceso se repite hasta que los centroides ya no cambien significativamente entre iteraciones o se alcance un número predeterminado de iteraciones.

**Pregunta 5:**

Enuncie al menos tres diferencias entre el análisis factorial y el método
de componentes principales.

**Respuesta 5:**

Diferencias entre el análisis factorial y el método de componentes principales:

1. Enfoque: El análisis factorial busca identificar relaciones latentes entre variables observadas, mientras que el método de componentes principales se enfoca en reducir la dimensionalidad de los datos.

2. Modelo subyacente: El análisis factorial se basa en la idea de que las variables observadas son indicadores imperfectos de constructos latentes, mientras que el método de componentes principales no asume una estructura latente subyacente.

3. Interpretación: En el análisis factorial, los factores son interpretados en términos de las variables observadas, mientras que en el método de componentes principales, los componentes son combinaciones lineales de las variables originales y pueden no tener una interpretación directa.

**Pregunta 6:**

¿Qué significa que el método de clusterización sea jerárquico?

**Respuesta 6:**

El significado puede ser entendido en 3 partes:

a) Estructura en capas: Un método de clusterización jerárquico organiza los grupos en una estructura de árbol, donde los clústeres se pueden subdividir en subclústeres de manera recursiva.

b) Relaciones de agrupamiento: La jerarquía muestra las relaciones de similitud entre los clústeres en diferentes niveles, lo que permite entender cómo se agrupan los datos a diferentes niveles de detalle.

c) Visualización: Permite visualizar la estructura de agrupamiento en forma de dendrograma, lo que facilita la comprensión de las relaciones entre los clústeres y la identificación de patrones en los datos.

**Usando los resultados anteriores clusterice el comportamiento de sus clientes:**

```python
# El área de marketing solicitó que el número de clusters sea no mayor a 8.
# Encuentre el número óptimo de clusters considerando esa restricción.

# Usamos WSS como en la documentación.

# Lista para almacenar los valores de WSS
wss_values = []

# Número máximo de clusters
max_clusters = 8

# Iterar sobre diferentes valores de k
for k in range(1, max_clusters + 1):
    # Inicializar y ajustar el modelo K-means
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data_filtrada_escalada)
    # Calcular la suma de los cuadrados dentro del grupo (WSS) y añadirla a la lista
    wss_values.append(kmeans.inertia_)

# Graficar el método del codo utilizando WSS
plt.plot(range(1, max_clusters + 1), wss_values, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Suma de cuadrados dentro del grupo (WSS)')
plt.title('Método del codo utilizando WSS para encontrar el número óptimo de clusters')
plt.show()
```

Se observa que a partir del número de clusters 8 la curva comienza a aplanarse, entonces considero que 8 es el número óptimo de clusters para el conjunto de datos.

```python
# Seleccionar las columnas relevantes de comportamiento de compra
comportamiento_compra_columnas = ['MntWines', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
                                  'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
                                  'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

df_comportamiento_compra = df_evaluacion_mkt_campaign[comportamiento_compra_columnas]

# Normalizar los datos
scaler = StandardScaler()
data_compra_escalada = scaler.fit_transform(df_comportamiento_compra)

# Definir el número de clusters
num_clusters = 8

# Inicializar el modelo KMeans
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Ajustar el modelo a los datos
kmeans.fit(data_compra_escalada)

# Obtener los centroides de cada cluster
centroides = kmeans.cluster_centers_

# Convertir los centroides a un DataFrame
centroides_df = pd.DataFrame(centroides, columns=comportamiento_compra_columnas)

# Imprimir los centroides
print("Centroides de cada cluster:")
print(centroides_df)
```

**Interpretación de los resultados:**

Estos centroides representan los valores medios de las variables en cada cluster.

Por ejemplo, en el cluster 0, los clientes tienen un gasto promedio en vinos (MntWines) de aproximadamente 0.909 desviaciones estándar por encima de la media, mientras que en el cluster 1, los clientes tienen un gasto promedio en carnes (MntMeatProducts) de aproximadamente 0.654 desviaciones estándar por debajo de la media.

**Nota importante:**

Si el valor es positivo, indica que, en promedio, los clientes en ese cluster tienden a gastar más en esa categoría de productos en comparación con el total de clientes.

Y si el valor es negativo en este contexto indica que, en promedio, los clientes en ese cluster tienden a gastar menos en esa categoría de productos en comparación con el total de clientes.

**Graficamos los Centroides:**

**Silhouette Score:**

```python
# Lista para almacenar los valores de Silhouette Score
silhouette_scores = []

# Número máximo de clusters
max_clusters = 8

# Iterar sobre diferentes valores de k
for k in range(2, max_clusters + 1):
    # Inicializar y ajustar el modelo K-means
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data_filtrada_escalada)
    # Calcular el Silhouette Score y añadirlo a la lista
    silhouette_scores.append(silhouette_score(data_filtrada_escalada, kmeans.labels_))

# Graficar el Silhouette Score en función del número de clusters
plt.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score en función del número de clusters')
plt.show()
```

**Davies_bouldin:**

```python
# Lista para almacenar los valores de Davies-Bouldin Index
davies_bouldin_scores = []

# Número máximo de clusters
max_clusters = 8

# Iterar sobre diferentes valores de k
for k in range(2, max_clusters + 1):
    # Inicializar y ajustar el modelo K-means
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data_filtrada_escalada)
    # Calcular el Davies-Bouldin Index y añadirlo a la lista
    davies_bouldin_scores.append(davies_bouldin_score(data_filtrada_escalada, kmeans.labels_))

# Graficar el Davies-Bouldin Index en función del número de clusters
plt.plot(range(2, max_clusters + 1), davies_bouldin_scores, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Davies-Bouldin Index')
plt.title('Davies-Bouldin Index en función del número de clusters')
plt.show()
```
